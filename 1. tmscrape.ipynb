```python
from bs4 import BeautifulSoup
from requests import get
import time
import random

url_base = 'https://www.transfermarkt.com/europameisterschaft-2024/abstellungverein/pokalwettbewerb/EM24/page/'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36 Edg/126.0.0.0',
}

teams = []
page_count = 1

while page_count <= 9:
    url = url_base + str(page_count)
    print(f'Requesting URL: {url}')
    response = get(url, headers=headers)

    if response.status_code != 200:
        print(f"Error: Failed to retrieve the page. Status code: {response.status_code}")
        break

    html_soup = BeautifulSoup(response.content, 'html.parser')
    teams_data = html_soup.find_all('tr', class_=["odd", "even"])

    if teams_data:
        teams.extend(teams_data)
        value = random.random()
        random_time = 1 + (value * (9 - 5))
        print(f'Time: {random_time} seconds')
        time.sleep(random_time)
    else:
        print('empty')
        break

    page_count += 1

print(f'Total teams found: {len(teams)}')

# To avoid IndexError, check if teams list is not empty
if len(teams) > 0:
    print(teams[0])

print()
total_teams = len(teams)
count = 0

while count < min(250, total_teams):
    html_soup = teams[count]

    # Extracting relevant information from the row
    team_name = html_soup.find('td', class_="links no-border-links hauptlink").text.strip()
    country = html_soup.find('td', class_="").find('img')['title']
    league = html_soup.find('td', class_="").find('a').text.strip()
    player_count = html_soup.find_all('td')[3].find('a').text.strip()
    total_value = html_soup.find_all('td', class_="rechts")[0].text.strip()
    average_value = html_soup.find_all('td', class_="rechts")[1].text.strip()

    print(f"{team_name}, {country}, {league}, {player_count}, {total_value}, {average_value}")
    count += 1
```

```python
from bs4 import BeautifulSoup
from requests import get
import time
import random

url_base = 'https://www.transfermarkt.com/copa-america-2024/abstellungverein/pokalwettbewerb/CAM4/page/'
headers = {
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/126.0.0.0 Safari/537.36 Edg/126.0.0.0',
}

teams = []
page_count = 1

while page_count <= 10:
    url = url_base + str(page_count)
    print(f'Requesting URL: {url}')
    response = get(url, headers=headers)

    if response.status_code != 200:
        print(f"Error: Failed to retrieve the page. Status code: {response.status_code}")
        break

    html_soup = BeautifulSoup(response.content, 'html.parser')
    teams_data = html_soup.find_all('tr', class_=["odd", "even"])

    if teams_data:
        teams.extend(teams_data)
        value = random.random()
        random_time = 1 + (value * (9 - 5))
        print(f'Time: {random_time} seconds')
        time.sleep(random_time)
    else:
        print('empty')
        break

    page_count += 1

print(f'Total teams found: {len(teams)}')

# To avoid IndexError, check if teams list is not empty
if len(teams) > 0:
    print(teams[0])

print()
total_teams = len(teams)
count = 0

while count < min(250, total_teams):
    html_soup = teams[count]

    # Extracting relevant information from the row with checks for missing elements
    team_name_elem = html_soup.find('td', class_="links no-border-links hauptlink")
    country_elem = html_soup.find('td', class_="").find('img') if html_soup.find('td', class_="") else None
    league_elem = html_soup.find('td', class_="").find('a') if html_soup.find('td', class_="") else None
    player_count_elem = html_soup.find_all('td')[3] if len(html_soup.find_all('td')) > 3 else None
    total_value_elems = html_soup.find_all('td', class_="rechts")

    if not team_name_elem or not country_elem or not player_count_elem or len(total_value_elems) < 2:
        print(f"Skipping row {count} due to missing elements")
        count += 1
        continue

    team_name = team_name_elem.text.strip()
    country = country_elem['title']
    league = league_elem.text.strip() if league_elem else "N/A"
    player_count = player_count_elem.find('a').text.strip() if player_count_elem.find('a') else "N/A"
    total_value = total_value_elems[0].text.strip()
    average_value = total_value_elems[1].text.strip()

    print(f"{team_name}, {country}, {league}, {player_count}, {total_value}, {average_value}")
    count += 1
```
